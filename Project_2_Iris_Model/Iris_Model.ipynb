## Importing libraries:

```python
import pandas as pd
import numpy as np
import matplotlib as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
from sklearn.preprocessing import LabelEncoder
```

## Load dataset:

```python
df=pd.read_csv(r"Data/Iris.csv")
```

```python
df.head(10)
```

#### Encoding values:
- Encoding species column values beacause model cann't understand string values we have to convert this to numeric values.

```python
le=LabelEncoder()   # instance of class labelencoder()
df["Species_en"]=le.fit_transform(df["Species"]) 
# Works only before dropping 'Species'   
```

```python
df = df.drop(columns="Species")  
df = df.drop(columns="Id")
# Remove categorical column before training
```

```python
df.head()  # Verify encoding result
```

## Split Data:

#### Split Data into input variables and target variables:

```python
x = df.drop(columns="Species_en")  # x = input variables or feature
y = df["Species_en"] # y = target variables or labeled
```

#### Split into train and test data:

```python
x_train,x_test, y_train, y_test =train_test_split(x,y,test_size=0.2,random_state=42)
```

## Training model:

```python
model = LogisticRegression(max_iter=1000)   # model instance of logisticRegression
```

```python
model.fit(x_train,y_train)   # model training on x_train,y_train
```

## Checking Accuracy:

```python
y_pred = model.predict(x_test)   # find predictions and store in variables. mean we find result of model on x test.
```

```python
accuracy = accuracy_score(y_test, y_pred)
```

```python
print(accuracy)   # accuracy is 1.0 mean 100% accurate.
```

## Prediction:

```python
print(x_test)
```

```python
model.predict([[6.1,2.8,4.7,1.2]])  # Predict class using trained model for a single input sample
```

```python
sample1 =pd.DataFrame([[6.1,2.8,4.7,1.2]], columns=[ 'SepalLengthCm', 'SepalWidthCm' , 'PetalLengthCm' , 'PetalWidthCm'])
# Create a sample input DataFrame with feature names for prediction

model.predict(sample1)
```

## Comparison of Actual Vs Prediction:

```python
result= pd.DataFrame({
    "actual": y_test,
    "predict": y_pred
})
# Create a DataFrame to compare actual values with model predictions side by side

print(result)   
```

## Conclusion
- From accuracy evaluation and comparison of actual vs predicted values, we conclude that the model is successfully trained.

## Save the Model:
- After training model, we don’t want to train it again and again.
- So we save it once, then later we can use it anytime.
- For saving model we have to import joblib.
- Joblib library use fro saving models.

```python
import pickle
```

```python
with open("model.pkl","wb") as f:
    pickle.dump(model, f) # dump is method of joblib which save model as model.pkl file
```
